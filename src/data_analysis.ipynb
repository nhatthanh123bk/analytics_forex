```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('Forex predict using linear regression')\
        .config('hive.metastore.uris', 'thrift://10.121.31.83:9083')\
        .enableHiveSupport()\
        .getOrCreate()
```

### Load data from Hive

```python
df_forex = spark.sql("SELECT * FROM forex.2012_m1_eurusd")
df_forex = df_forex.drop('volume')
df_forex.registerTempTable("forex")
```

```python
forex_29_31_12 = spark.sql("SELECT * FROM forex WHERE date BETWEEN '2012.12.24' AND '2012.12.28'").show()
forex_29_31_12.show()
```
